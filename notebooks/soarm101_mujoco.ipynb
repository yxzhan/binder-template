# SOARM101 MuJoCo Simulation

Launch the SO-101 robot arm in MuJoCo with ROS 2 `/joint_states` bridge.


<button data-commandlinker-command="notebook:run-all-cells" class="jupyter-button" style="color: #fff;background-color: #1976d2;">Launch Interface</button>

```python
from IPython.display import display, HTML, IFrame
import subprocess
import ipywidgets as widgets
from utils import display_desktop, get_base_url

sim_process = None

output = widgets.Output()

def start_simulation(_):
    global sim_process
    with output:
        output.clear_output()
        if sim_process is not None and sim_process.poll() is None:
            print("Simulation is already running.")
            return
        sim_process = subprocess.Popen(["python", "soarm101_mujoco/simulate.py", "--ros"])
        print("Simulation started.")

def stop_simulation(_):
    global sim_process
    with output:
        output.clear_output()
        if sim_process is None or sim_process.poll() is not None:
            print("Simulation is not running.")
            return
        sim_process.terminate()
        sim_process.wait(timeout=5)
        sim_process = None
        print("Simulation stopped.")

start_btn = widgets.Button(description="Start Simulation", button_style="success")
stop_btn = widgets.Button(description="Stop Simulation", button_style="danger")

start_btn.on_click(start_simulation)
stop_btn.on_click(stop_simulation)

display(widgets.HBox([start_btn, stop_btn]))

url_prefix = get_base_url()
ros2ws = f"{url_prefix}/proxy/9090".replace("http", "ws")
web_url = f"{url_prefix}/proxy/3000/?ros2ws={ros2ws}"
desktop_url = f"{url_prefix}/desktop"

display(HTML(f"""                                                                                                                                                   
  <div style="display:flex; gap:4px; height:700px;">        
    <iframe src="{web_url}" style="flex:1; border:0;"></iframe>                                                                                                       
    <iframe src="{desktop_url}" style="flex:1; border:0;"></iframe>                                                                                                   
  </div>                                                                                                                                                              
  """))  
display(output)
```
